```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
historia_temperaturas = pd.read_csv('/home/ec2-user/clubes_de_ciencia/Dia_4/historia_temperaturas.csv')
produccion_petroleo = pd.read_csv('/home/ec2-user/clubes_de_ciencia/Dia_4/produccion_petrolio.csv')
produccion_petroleo.Date = pd.to_datetime(produccion_petroleo.Date)
historia_temperaturas.dt = pd.to_datetime(historia_temperaturas.dt)
```

```python
plt.figure(figsize=(8,8))
plt.plot(historia_temperaturas.dt,historia_temperaturas.temperatura_promedio)
plt.xlabel('Ao',fontsize = 14, fontweight='bold')
plt.ylabel('Temperatura($c^{0}$)', fontsize = 14,fontweight='bold')
plt.title('Temperatura promedio anual Mundial',fontsize = 16,fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('temperatura_promedio_anual')
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.plot(produccion_petroleo.Date,produccion_petroleo.Volume)
plt.xlabel('Ao',fontsize = 14, fontweight='bold')
plt.ylabel('Volumen', fontsize = 14,fontweight='bold')
plt.title('Produccin petroleo anual Mundial',fontsize = 16,fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('produccion_petroleo')
plt.show()
```

```python
x = produccion_petroleo[produccion_petroleo.Date.dt.year <= 2015].Volume
y = historia_temperaturas[historia_temperaturas.dt.dt.year >= 1983].temperatura_promedio
```

```python
plt.figure(figsize=(9,9))
plt.scatter(x,y,c='blue')
plt.xlabel('Produccin Petroleo',fontsize = 14, fontweight='bold')
plt.ylabel('Temperatura promedio($c^{0}$)', fontsize = 14,fontweight='bold')
plt.title('Produccin petroleo anual Mundial',fontsize = 16,fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('temperatura_petroleo')
plt.show()
```
